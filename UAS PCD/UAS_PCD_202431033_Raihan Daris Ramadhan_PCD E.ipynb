```python
#import library yg digunakan

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Memuat Citra Asli
path_gambar = "1001080188.jpg"
citra_asli = cv2.imread(path_gambar)
citra_asli = cv2.cvtColor(citra_asli, cv2.COLOR_BGR2RGB)

tinggi, lebar = citra_asli.shape[:2]

plt.imshow(citra_asli)
plt.title("Citra Asli")
plt.show()
```

```python
#rotasi Citra

sudut_rotasi = 25
titik_pusat = (lebar // 2, tinggi // 2)
matriks_rotasi = cv2.getRotationMatrix2D(titik_pusat, sudut_rotasi, 1.0)
citra_rotated = cv2.warpAffine(citra_asli, matriks_rotasi, (lebar, tinggi))

plt.imshow(citra_rotated)
plt.title("After Rotated")
plt.show()
```

```python
#Penskalaan Citra
skala = 0.5
lebar_baru = int(lebar * skala)
tinggi_baru = int(tinggi * skala)
citra_resized = cv2.resize(citra_asli, (lebar_baru, tinggi_baru), interpolation=cv2.INTER_AREA)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(citra_asli)
plt.title("Citra Asli")

plt.subplot(1, 2, 2)
plt.imshow(citra_resized)
plt.title("After Resized")

plt.show()
```

```python
#Pemotongan Citra 

y1, y2 = int(tinggi * 0.35), int(tinggi * 0.65)
x1, x2 = int(lebar * 0.30), int(lebar * 0.70)
citra_cropped = citra_asli[y1:y2, x1:x2]

plt.imshow(citra_cropped)
plt.title("After Cropped")
plt.show()
```

```python
#Pencerminan Citra

citra_flipped = cv2.flip(citra_asli, 1)

plt.imshow(citra_flipped)
plt.title("After Flipped")
plt.show()
```

```python
#Translasi/ pergeseran citra

tx, ty = 80, 50
matriks_translasi = np.float32([[1, 0, tx], [0, 1, ty]])
citra_translated = cv2.warpAffine(citra_asli, matriks_translasi, (lebar, tinggi))

plt.imshow(citra_translated)
plt.title("After Translated")
plt.show()
```

```python
#Verif Ukuran Citra dan Matriks Transformasi

print('Ukuran Citra Asli      :', citra_asli.shape)
print('Ukuran Citra Rotated   :', citra_rotated.shape)
print('Ukuran Citra Resized   :', citra_resized.shape)
print('Ukuran Citra Cropped   :', citra_cropped.shape)
print('Ukuran Citra Flipped   :', citra_flipped.shape)
print('Ukuran Citra Translated:', citra_translated.shape)

print()
print('Matriks Rotasi (2x3):')
print(matriks_rotasi)

print('Matriks Translasi (2x3):')
print(matriks_translasi)
```
